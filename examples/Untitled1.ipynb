```python
import numpy as np
import pandas as pd
from zsmash.smash import Smash
```

We can datasmash symbolic data (i.e. data consisting of only 0s and 1s) using the following:

```python
data = pd.read_csv("./data/crime_short.dat", sep=" ", header=None)
smash = Smash(data=data, data_type='symbolic', data_dir='row')
pd.DataFrame(smash.run())
```

We can also datasmash continuous data by first converting the data to symbolic.

```python
data = pd.read_csv("./data/YHOOMSFT.dat", sep=" ", header=None)
display(data)
data = data.apply(lambda row: row - row.shift(1), axis=1).dropna(axis=1)
data = data.apply(lambda row: row.apply(lambda val: 1 if val > 0 else 0), axis=1)
display(data)

smash = Smash(
    data=data,
    data_type='symbolic',
    data_dir='row'
)
pd.DataFrame(smash.run())
```
